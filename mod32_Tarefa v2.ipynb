# Probabilidade - Tarefa

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
media = 15  # Média
desvio_padrao = 0.1  # Desvio padrão
limite_superior = 15.2  # Limite superior

# Calculando a probabilidade usando a distribuição normal acumulada (CDF)
probabilidade = 1 - stats.norm.cdf(limite_superior, loc=media, scale=desvio_padrao)

print(f"A probabilidade de um pacote ter mais que 15.2 kg é {probabilidade:.2%}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
media = 15  # Média
desvio_padrao = 0.1  # Desvio padrão
percentil = 0.95  # Percentil desejado

# Calculando o percentil usando a distribuição normal
percentil_valor = stats.norm.ppf(percentil, loc=media, scale=desvio_padrao)

print(f"O percentil de ordem 95% para o peso dos pacotes é {percentil_valor:.2f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
#Y=np.linspace(-6, 6, 200)
# Seu código começa aqui

# Definindo os parâmetros das distribuições
media_norm = 0
desvio_padrao_norm = 1
graus_liberdade_t = 5

# Valores contidos no objeto Y
Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades para Y1 (Normal Padrão)
probabilidades_Y1 = [stats.norm.cdf(valor, loc=media_norm, scale=desvio_padrao_norm) for valor in Y]

# Calculando as probabilidades para Y2 (t-student)
probabilidades_Y2 = [stats.t.cdf(valor, df=graus_liberdade_t) for valor in Y]

# Printando os resultados
for i, valor in enumerate(Y):
    print(f"Para Y1: A probabilidade de ser menor que {valor:.2f} é {probabilidades_Y1[i]:.4f}")
    print(f"Para Y2: A probabilidade de ser menor que {valor:.2f} é {probabilidades_Y2[i]:.4f}")
    print("="*30)
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Definindo os parâmetros das distribuições
media_norm = 0
desvio_padrao_norm = 1
graus_liberdade_t = 5

# Valores contidos no objeto Y
Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades para Y1 (Normal Padrão)
probabilidades_Y1 = [stats.norm.cdf(valor, loc=media_norm, scale=desvio_padrao_norm) for valor in Y]

# Calculando as probabilidades para Y2 (t-student)
probabilidades_Y2 = [stats.t.cdf(valor, df=graus_liberdade_t) for valor in Y]

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando as probabilidades para Y1 e Y2
ax.plot(Y, probabilidades_Y1, label='Y1 - Normal Padrão')
ax.plot(Y, probabilidades_Y2, label='Y2 - t-student')

# Configurando os eixos e o título
ax.set_xlabel('Valores de Y')
ax.set_ylabel('Probabilidade')
ax.set_title('Probabilidade acumulada para Y1 e Y2')
ax.legend()

# Exibindo o gráfico
plt.show()
```
